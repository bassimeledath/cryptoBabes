```python
import pandas as pd
import numpy as np
import math
```

```python
data = pd.read_csv('original_data.csv')
print (data.head())
```

```python
from math import log
fisher_transform = []
for x in data['price']:
    fisher = 0.5*log(abs((1+x)/(1-x)))
    fisher_transform.append(fisher)

data['fisher_transform'] = pd.Series(fisher_transform)
```
