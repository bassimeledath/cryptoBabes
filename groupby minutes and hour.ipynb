```python
import pandas as pd
bitdata = pd.read_csv('preprocessed1.csv')
print(bitdata)
```

```python
import numpy as np
#bitdata.groupby?
#bitdata.agg?

for group_name, group in bitdata.groupby(['date','hour_minute']):
    print(group_name)
    print(np.average(group['price']))
    print(np.max(group['price']))
    print(np.min(group['price']))
```

```python
max_list = bitdata.groupby(['date','hours','minutes']).agg({'price': np.max})
max_list.columns=['max_price']
```

```python
min_list = bitdata.set_index('hour_minute').groupby(['date','hours','minutes']).agg({'price': np.min})
min_list.columns=['min_price']
```

```python
average_list = bitdata.set_index('hour_minute').groupby(['date','hours','minutes']).agg({'price': np.average})
average_list.columns = ['average_price']
```

```python
result = max_list
result = result.join(min_list['min_price'])
result = result.join(average_list['average_price'])
print(result)
result.to_csv('group_by_minute.csv',index=False)
```

```python
max_list2 = bitdata.groupby(['date','hours',]).agg({'price': np.max})
max_list2.columns=['max_price']
min_list2 = bitdata.groupby(['date','hours',]).agg({'price': np.min})
min_list2.columns=['min_price']
avg_list2 = bitdata.groupby(['date','hours',]).agg({'price': np.average})
avg_list2.columns=['avg_price']
result = max_list2
result = result.join(min_list2['min_price'])
result = result.join(avg_list2['avg_price'])
print(result)
result.to_csv('group_by_hour.csv',index=False)
```
