```python
import pandas as pd
bitdata = pd.read_csv('/Users/yuqizhang/my-python-project/cryptoBabes/Bitfinex_BTCEUR_trades_2018_02_02.csv')
print(bitdata)
```

```python
import matplotlib.pyplot as plt
import dateandtime
```

```python
def change_date(row):
    data = row['date']
    return pd.Series({'date':dateandtime.convert_dates(data)})

new_date = bitdata.apply(change_date, axis = 1)


print(new_date)
```

```python
type(new_date)
```

```python
new_bitdata = bitdata.drop(columns=['exchange','symbol','date'])
new_bitdata = new_bitdata.join(new_date)


```

```python
new_bitdata.to_csv('datawithnewdate.csv',index=False)
```
