```python
import pandas as pd
file1 = pd.read_csv('group_by_minute.csv')
#print (file1.head())
file2 = pd.read_csv('open&close_minute.csv')
file1 = file1.join(file2[['open','close']])
file1.columns=['hour_minute','high','low','average','open','close']
print(file1)
file1.to_csv('four_feature_minute.csv',index=False)
```

```python
import pandas as pd
file1 = pd.read_csv('group_by_hour.csv')
file2 = pd.read_csv('open&close_hour.csv')
file1 = file1.join(file2[['open','close']])
file1.columns=['hours','high','low','average','open','close']
print(file1)
file1.to_csv('four_feature_hour.csv',index=False)
```
